```python
import json
from datetime import datetime, timedelta

with open('processed-hurdat2.json', 'r') as f:
    hurdat2data = json.load(f)
# Now you can work with the data loaded from the JSON file.
# Example: print the keys of the first item in the data list
if isinstance(hurdat2data, list) and len(hurdat2data) > 0:
    print(hurdat2data)
else:
  print("The JSON data is empty or not a list.")

focusYear = '2020'
yearsInHurricane = []

for i in hurdat2data:
    if i['storm_id'] == focusYear:
        yearsInHurricane.extend(i['observations'])

yearsInHurricane.sort(key=lambda x: datetime.strptime(f"{x['date']} {x['time']}", '%Y-%m-%d %H:%M UTC'))

firstObservationDate = datetime.strptime(yearsInHurricane[0]['date'], '%Y-%m-%d')
lastObservationDate = datetime.strptime(yearsInHurricane[-1]['date'], '%Y-%m-%d')
print(firstObservationDate)
print(lastObservationDate)

months = []
current_date = firstObservationDate
while current_date <= lastObservationDate:
    months.append(current_date.strftime("%m"))  # Extract the month as a string
    current_date += timedelta(days=31)  # Move to the next month (approximately)
    current_date = current_date.replace(day=1)  # Reset to the first day of the next month

# Remove duplicate months and ensure correct ordering
all_months = sorted(list(set(months)))

# Zero-pad the month strings to two places
zero_padded_months = [month.zfill(2) for month in all_months]

zero_padded_months
```

```python
import xarray as xr

# Load the NetCDF dataset
mslAndSST = xr.open_dataset('data_stream-oper_stepType-instant.nc')
pressure = xr.open_dataset('pressure-data.nc')

print(mslAndSST)
print(pressure)
```

```python
# Round latitude and longitude to the nearest degree
mslAndSST = mslAndSST.assign_coords(
    latitude=mslAndSST.latitude.round(0),
    longitude=mslAndSST.longitude.round(0)
)
# Group by the rounded coordinates and compute the mean temperature
mslAndSST = mslAndSST.groupby(['latitude', 'longitude']).mean()
print(mslAndSST)
```

```python
def dropRandomColumns(ds):
    if 'expver' in ds:
        ds = ds.drop_vars('expver')
        print("Column 'expver' dropped successfully.")
    else:
        print("Column 'expver' not found in the dataset.")

    if "number" in ds.keys():
        ds = ds.drop_vars('number')
        print("Column 'number' dropped successfully.")
    else:
        print("Column 'number' not found in the dataset.")

    return ds

mslAndSST = dropRandomColumns(mslAndSST)
mslAndSST
```

```python
mslAndSST = mslAndSST.to_dataframe()
mslAndSST = mslAndSST.reset_index()
mslAndSST
```

```python
# Round the 'msl' column to the nearest whole number
mslAndSST['msl'] = mslAndSST['msl'].round(0)
mslAndSST['sst'] = mslAndSST['sst'].round(0)
mslAndSST
```

```python
mslAndSST.to_csv('/content/processed/msl_and_sst.csv', index=False)
```

```python
def roundLatLong(ds):
    # Round latitude and longitude to the nearest degree
    ds = ds.assign_coords(
        latitude=ds.latitude.round(0),
        longitude=ds.longitude.round(0)
    )
    # Group by the rounded coordinates and compute the mean temperature
    ds = ds.groupby(['latitude', 'longitude']).mean()
    return ds

pressure = roundLatLong(pressure)
pressure
```

```python
pressure = dropRandomColumns(pressure)
pressure
```

```python
import pandas as pd
import numpy as np

try:
    u_850 = pressure['u'].sel(pressure_level=850)
    u_200 = pressure['u'].sel(pressure_level=200)
    v_850 = pressure['v'].sel(pressure_level=850)
    v_200 = pressure['v'].sel(pressure_level=200)

    vertical_wind_shear = np.sqrt((u_850 - u_200)**2 + (v_850 - v_200)**2)

    pressure['vertical_wind_shear'] = vertical_wind_shear

except KeyError as e:
    print(f"Error: Key {e} not found in the dataset. Check variable and level names.")

except ValueError as e:
    print(f"Error processing data: {e}")

pressure
```

```python
pressure_700 = pressure['r'].sel(pressure_level=700)
pressure['relative_humidity'] = pressure_700

pressure
```

```python
vorticity = pressure['vo'].sel(pressure_level=850)
pressure['vorticity'] = vorticity
pressure
```

```python
def dropColumns(ds, cols=['expver', 'number']):
    for i in cols:
        if i in ds:
            ds = ds.drop_vars(i)
            print(f"Column '{i}' dropped successfully.")
        else:
            print(f"Column '{i}' not found in the dataset.")

    return ds

pressure = dropColumns(pressure, cols=['r', 'u', 'v', 'vo'])
pressure
```

```python
pressure = pressure.to_dataframe()
pressure = pressure.reset_index()
pressure
```

```python
pressure = pressure.drop(columns=['pressure_level'])
pressure
```

```python
pressure.drop_duplicates(inplace=True)
pressure
```
